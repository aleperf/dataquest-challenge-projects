The Posts table contains data about 8 different kind of post, the most important in volume are questions and answers, the main columns are:

- Id: an indentification number for the post
- PostTypeId: a number identifying the type of post (if answer, or question for example).
- CreationDate: the date of the post
- Score
- ViewCount: how many views the post received
- Tags: the tags used to mark the post
- AnswerCount: the number of answers to the post (if it is a question)
- FavoriteCount: the number of times the post was favored by other users (only questions post can be favored).

### Popular Tags

We will find the most popular tags, considering as such the most used tags in posts and the most viewed tags.

```python
all_tags = set()
```

```python
for tags in posts_2019["Tags"]:
    current_tags = tags.split(",")
    for element in current_tags:
        element = element.replace(".","\.")
        element = element.replace("+", "\+")
        all_tags.add(element)
```

```python
all_tags
```

```python
popular = {}
most_viewed = {}
```

```python
for element in all_tags:
    contained = posts_2019["Tags"].str.contains(element)
    number = contained.sum()
    viewed = posts_2019[contained]["ViewCount"].sum()
    name = element.strip("\b")
    popular[name] = number
    most_viewed[name] = viewed
```

### Interest in Deep Learnign Across Time

We have seen data about the year 2019, we'll now analyze data about all the questions in Data Science Stack Exchange until 2019, in order to understand the interest in Deep Learning across time, considering some of the popular tags we have seen in the previous section.

The questions considered in the all_questions dataset have a date between May 13th, 2014 and January 19th, 2019. <br>
In order to evaluate the interest in deep learning across time, I'll consider a timeframe of six months, looking for the following keyword related to deep-learning: deep-learning, neural-network, keras, tensorflow, image-classification, which are between the most popular or most viewed in 2019.
I'll count the number of posts with at least one tag related to deep learning. <br>
This isn't a comprehensive survey about all deep learning arguments considered in StackExchange, but just an attempt to evaluate a variation in interest about the general argument of deep learning, so real data are probable more in favor of deep learning because I won't consider many topics related to this field.

```python
all_questions = pd.read_csv("all_questions.csv")
```

```python
all_questions.info()
```

```python
all_questions.head()
```

```python
all_questions["CreationDate"] = pd.to_datetime(all_questions["CreationDate"])
```

```python
all_questions["CreationDate"].min()
```

```python
all_questions["CreationDate"].max()
```

```python
all_questions["Tags"] = all_questions["Tags"].str.replace("<", "\b").str.replace(">", ",\b").str.replace(",\b$", "\b")
```

```python
keywords = ["keras", "deep-learning", "neural-network", "tensorflow","image-classification"]
```

```python
import datetime as dt
from datetime import datetime
```

```python
from dateutil.relativedelta import relativedelta
```

```python
start_date = dt.datetime(2014, 5, 13)
```

```python
six_months = relativedelta(months=+6)
```

```python
dl_interest = []
```

```python
end_dates = []
```

```python
for i in range(0, 10):
    start = start_date + (six_months * i)
    end = start + six_months
    end_dates.append(end)
    mask = (all_questions["CreationDate"] >= start) &\
    (all_questions["CreationDate"] < end)
    df = all_questions[mask]
    df_interest = pd.DataFrame(columns=keywords)
    for arg in keywords:
        df_interest[arg] = df["Tags"].str.contains(arg)
    dl_interest.append(df_interest)    
```

```python
len(dl_interest)
```

```python
#example of one of the six months period dataframes
dl_interest[9].head()
```

```python
total_interest = []
keras_interest = []
deep_learning_interest = []
neural_network_interest = []
tensorflow_interest = []
image_classification_interest = []
```

```python
#I want to count a post only one time as deep learnign related, so if a post contains more 
# than one keyword, it should be counted just as one, zero if not related.
def convert_to_interest(total_sum):
    if total_sum > 0:
        return 1
    return 0
```

```python
for df in dl_interest:
    df["total_interest"] = df.sum(axis=1)
    df["total_interest"]=  df["total_interest"].apply(convert_to_interest)
    total_interest.append(df["total_interest"].sum())
    keras_interest.append(df["keras"].sum())
    deep_learning_interest.append(df["deep-learning"].sum())
    neural_network_interest.append(df["neural-network"].sum())
    tensorflow_interest.append(df["tensorflow"].sum())
    image_classification_interest.append((df["image-classification"].sum()))
```

```python
date_labels = []
for date in end_dates:
    date_labels.append(date.strftime("%Y-%m"))
```

```python
positions = np.arange(len(dl_interest)) + 1
```

```python
plt.figure (figsize=(10,5))
plt.bar(positions, total_interest, align = "center")
plt.xticks(positions, date_labels, rotation=70);
plt.title("Interest in Deep Learning Across 2014-2019")
plt.ylabel("Number of Posts Related to Deep Learning")
plt.xlabel("End of Semester")
plt.show();
```

The plot above shows the number of questions containing at least one of the arguments between the keywords and it is clear that the interest in deep learning is growing more and more across time, being at peak in the first semester of 2019. <br>
In 2014, the starting time of our dataset, deep learning was less considered.

```python
plt.figure (figsize=(10,5))
plt.bar(positions, keras_interest, align = "center")
plt.xticks(positions, date_labels, rotation=70);
plt.title("Interest in Keras Across 2014-2019")
plt.ylabel("Number of Posts Related to Keras")
plt.xlabel("End of Semester")
plt.show();
```

Keras, a popular frameworkd for deep learning based on Tensorflow, was created in March 2015, so clearly there are no posts about Keras in the first semesters of our dataset, but starting from fourth semester, in 2016, the interest in Keras has grown similarly to the general interest in deep learning.

```python
plt.figure (figsize=(10,5))
plt.bar(positions, tensorflow_interest, align = "center")
plt.xticks(positions, date_labels, rotation=70);
plt.title("Interest in Tensorflow Across 2014-2019")
plt.ylabel("Number of Posts Related to Tensorflow")
plt.xlabel("End of Semester")
plt.show();
```

The interest in Tensorflow mirrors the one in Keras (and viceversa), the two arguments are in fact strictly related, Keras is based on Tensorflow.

```python
plt.figure (figsize=(10,5))
plt.bar(positions, neural_network_interest, align = "center")
plt.xticks(positions, date_labels, rotation=70);
plt.title("Interest in Neural Networks Across 2014-2019")
plt.ylabel("Number of Posts Related to Neural Networks")
plt.xlabel("End of Semester")
plt.show();
```

The interest in neural networks has grown over time since 2014, but peaked in the second to last semester (November 2018), slightly descending in the last semester (May 2019). 

```python
plt.figure (figsize=(10,5))
plt.bar(positions, image_classification_interest, align = "center")
plt.xticks(positions, date_labels, rotation=70);
plt.title("Interest in Image Classification Across 2014-2019")
plt.ylabel("Number of Posts Related to Image Classification")
plt.xlabel("End of Semester")
plt.show();
```

Image Classification is one of the most common problems in Deep Learning, so no surprisingly the interest in this topis has grown over time since the second semester of our dataset.

```python
plt.figure (figsize=(10,5))
plt.plot(end_dates, total_interest, label="Total Interest")
plt.plot(end_dates, keras_interest, label= "Keras")
plt.plot(end_dates, tensorflow_interest, label="Tensorflow")
plt.plot(end_dates, image_classification_interest, label="Image Classification")
plt.plot(end_dates, neural_network_interest, label="Neural Networks")
plt.title("Interest in Deep Learning on StackExchange Data Science in 2014-2019")
plt.ylabel("Number of Posts")
plt.xlabel("End of Semester")
plt.legend(loc="upper left")
plt.xticks(end_dates, date_labels, rotation=70);
```

Comparing all the arguments at once, it is clear there is trend of growing interest towards Deep Learning, with only a slightly flession in Neural Networks.
